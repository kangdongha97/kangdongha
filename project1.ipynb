```python
"""심장병 유무"""
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
heart = pd.read_csv('heart.csv') #1.데이터 형식전환
group_names = ['건강','심장병']

heart['state'], mybin = pd.cut(heart['target'], 2, labels=group_names, retbins=True) #2.cut사용 target으로 state컬럼을 만듦

heart['chol'] = (heart['chol']-heart['chol'].min())/(heart['chol'].max()-heart['chol'].min()) #5.정규화

heart = pd.concat([heart, pd.get_dummies(heart['state'])], axis=1, sort=False) #3.get_dummies
heart
```

```python
X = heart[['age','sex','cp','trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']]
feature_data = np.array(X)
y = heart['state']
label_data = np.array(y)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature_data, label_data, test_size=.5)
from sklearn import tree
clf = tree.DecisionTreeClassifier() #결정트리모델을 사용
clf.fit(X_train, y_train) #학습
predictions = clf.predict(X_test) #예측
print('예측:\n', predictions)
print('정답(y_test):\n', y_test)
print("정답률 : ", accuracy_score(y_test, predictions)) #6.정확도
```

```python
from scipy.spatial import distance
def euc(a,b) :
    return distance.euclidean(a, b)
class eucKNN():
    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train
    def predict(self, X_test):
        predictions = []
        for row in X_test:
            label = self.closest(row)
            predictions.append(label)
        return predictions
    def closest(self, row):
        best_dist = euc(row, self.X_train[0])
        best_index = 0
        for i in range(1, len(self.X_train)):
            dist = euc(row, self.X_train[i])
            if dist < best_dist:
                best_dist = dist
                best_index = 1
        return self.y_train[best_index]
X = heart[['age','sex','cp','trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']]
feature_data = np.array(X)
y = heart['state']
label_data = np.array(y)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature_data, label_data, test_size=.5)
clf = eucKNN() #유클리디언
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
from sklearn.metrics import accuracy_score
print('예측:\n', predictions)
print('정답(y_test):\n', y_test)
print("정답률 : ", accuracy_score(y_test, predictions))
```
